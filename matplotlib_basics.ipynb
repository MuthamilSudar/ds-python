 The autopct parameter in Matplotlib's plt.pie() function is used to format the values displayed on a pie chart to indicate the percentage of each wedge relative to the total.


 startangle 0 - degrees corresponds to the right-most side of the plot (3 o'clock position), and angles increase counterclockwise.
 startangle=90: Starts drawing the first wedge at the top of the plot (12 o'clock position), rotating counterclockwise.
 startangle=-45: Starts drawing the first wedge at a 45-degree angle clockwise from the right-most side of the plot.

```python
import matplotlib.pyplot as plt
categories = ['A', 'B', 'C', 'D', 'E']
values = [15, 30, 25, 20, 10]
plt.pie(values, labels=categories, autopct='%1.1f%%', startangle=90, colors=['orange', 'yellow', 'green', 'blue', 'purple'])
plt.title('Pie Chart')
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def f(x, y):
    return x**2 + y**2


x = np.linspace(-5, 5, 100)
# print(x)
y = np.linspace(-5, 5, 100)


X, Y = np.meshgrid(x, y)
Z = f(X, Y)

# Plot the contour graph
plt.contour(X, Y, Z,20,cmap='viridis')

plt.xlabel('X')
plt.ylabel('Y')
plt.title('Contour Plot of x^2 + y^2')
plt.colorbar(label='Function Value')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def f(x, y):
    return x**2 + y**2

x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)

X, Y = np.meshgrid(x, y)
Z = f(X, Y)

# Create subplots
plt.subplot(2, 1, 1)
plt.contourf(X, Y, Z, 20, cmap='viridis')  # Filled contour plot
plt.colorbar(label='Function Value')
plt.title('Filled Contour Plot of x^2 + y^2')

plt.subplot(2, 1, 2)
plt.contour(X, Y, Z, 20, cmap='viridis')  # Contour plot
plt.colorbar(label='Function Value')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Contour Plot of x^2 + y^2')

plt.tight_layout()
plt.show()
```

x represents the x-values.

y represents the y-values, generated randomly here.

dy represents the error or uncertainty in the y-values, generated randomly as well.

yerr=dy specifies that the error bars for each data point should be drawn using the values specified in the dy array.

fmt='o' specifies that the data points should be represented as circles ('o').

color='b' sets the color of the data points to blue.

ecolor='r' sets the color of the error bars to red.

capsize=5 sets the size of the caps at the end of each error bar to 5 points.

```python
import numpy as np
import matplotlib.pyplot as plt
x = np.arange(10)
y = np.random.randint(0, 10, size=10)
print(y)
dy = np.random.uniform(0.5, 1.5, size=10)
print(dy)
plt.errorbar(x, y, yerr=dy, fmt='o', color='b', ecolor='r', capsize=5)
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Error Bar Plot')
plt.grid(True)
plt.show()
```

 Hexagonal binning is another method for visualizing the distribution of two-dimensional data, particularly useful when dealing with a large number of data points.

```python
import numpy as np
import matplotlib.pyplot as plt
np.random.seed(0)
x = np.random.randn(1000)
y = np.random.randn(1000)
# Create the hexagonal binning plot
plt.hexbin(x, y, gridsize=30, cmap='viridis')
plt.colorbar(label='Count')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Hexagonal Binning')
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate some example data
np.random.seed(0)
data1 = np.random.randn(1000)
data2 = np.random.randn(1000)

# Plot histograms with different histtype options
plt.figure(figsize=(10, 4))

plt.subplot(1, 3, 1)
plt.hist([data1, data2], bins=20, histtype='bar', color=['blue', 'green'], alpha=0.7)
plt.title('bar')

plt.subplot(1, 3, 2)
plt.hist([data1, data2], bins=20, histtype='step', color=['blue', 'green'], alpha=0.7)
plt.title('step')

plt.subplot(1, 3, 3)
plt.hist([data1, data2], bins=20, histtype='stepfilled', color=['blue', 'green'], alpha=0.7)
plt.title('stepfilled')

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec


x = np.linspace(0, 10, 100)
y = np.sin(x)
data = np.random.randn(1000)
x_scatter = np.random.rand(50) * 10
y_scatter = np.random.rand(50) * 2

# Define the grid
grid = gridspec.GridSpec(2, 3)

# Create subplots using the grid specification
plt.subplot(grid[0, 0])
plt.bar(x, y, color='blue', alpha=0.7)
plt.title('Bar Plot')

plt.subplot(grid[0, 1:])
plt.scatter(x_scatter, y_scatter, color='red', alpha=0.7)
plt.title('Scatter Plot')

plt.subplot(grid[1, :2])
plt.plot(x, y, color='green')
plt.title('Line Plot')

plt.subplot(grid[1, 2])
plt.hist(data, bins=30, color='orange', alpha=0.7)
plt.title('Histogram')

# Adjust layout and display plot
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
x = np.linspace(0, 10, 100)
y = np.sin(x)
plt.plot(x, y, label='sin(x)')
# Add text annotation
plt.text(3, 0.5, 'Middle', fontsize=12, color='red', style='italic')
# Add arrow annotation
plt.annotate('Peak', xy=(np.pi/2, 1), xytext=(3, 0.8),
             arrowprops=dict(facecolor='black', arrowstyle='->'))
plt.xlabel('x')
plt.ylabel('y')
plt.title('Text and Annotation Example')
plt.legend()
plt.grid(True)
plt.show()
```

plt.text is used to add a text annotation at coordinates (3, 0.5). We specify the text, fontsize, color, and style of the text.

plt.annotate is used to add an arrow annotation pointing to the peak of the sine wave at (np.pi/2, 1).

We specify the text, the location of the arrow tip (xy), and the position of the text (xytext). Additionally, we customize the arrow properties using arrowprops.

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate some example data
x = np.arange(0, 10)
y = x ** 2

# Plot the data
plt.plot(x, y)

# Customize xticks and yticks
plt.xticks([0, 2, 4, 6, 8, 10])  # Customize x-axis ticks
plt.yticks([0, 20, 40, 60, 80, 100])  # Customize y-axis ticks

# Add grid
plt.grid(True)

plt.xlabel('X')
plt.ylabel('Y')
plt.title('Customized Ticks Example')

plt.show()
```
